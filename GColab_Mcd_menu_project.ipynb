Checking the relationship between Calories and Calories from Fat:
 - There is a strong relationship between Calories 
   and Calories with fat. Hence, food items with high caloric content tend to also have a high calories with fat content.

-  McDonalds foods have low trans fat which is 
   good.

-  The Smoothies and Shakes Menu Category had the 
   highest saturated fat content. Upon analysis, it was discovered that while Fruit shakes had no saturated fat. However, milk Shakes, Chocolate Shakes, Shamrock and McFlurry shakes had high saturated fat with McFlurry at the top of the list.

```python
plt.figure(figsize=(10, 4), dpi=100)

correlation = data['Calories'].corr(data['Calories from Fat'])
plt.scatter(data.Calories, data['Calories from Fat'], color='black')
plt.text(500,450,'r = {}'.format(round(correlation,2)))
plt.xlabel("Calories")
plt.ylabel("Calories from Fat")
plt.title("Relationship between Calories and Calories from Fat")
plt.show()
```

Finding food categories with the highest % Saturated Fat.

```python
data['saturated_cholesterol'] = data['Saturated Fat']/data['Cholesterol']*100
```

```python
saturated_cholesterol = data.groupby('Category')['saturated_cholesterol'].mean().dropna().nlargest(10)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
saturated_cholesterol.sort_values(ascending=False).plot.bar(color = 'red')
plt.title("Food categories with the highest % Saturated Fat")
plt.ylabel("Percentage")
plt.xlabel("Menu Category")
plt.xticks(rotation=0)
plt.show();
```

Visualization of the iron and calcium contents in the Mcdonald's menu.
 - Mcdonald's beverages are very low in Calcium and Iron.
 - Menu categories rich in Animal protein (Beef & Pork, Chicken & Fish) are rich in both Iron and Calcium.
 - Shakes (fruit shakes, milk shakes etc.), coffee and tea are rich in calcium but low in iron

```python
iron = data.groupby('Category')['Iron (% Daily Value)'].mean()
calcium = data.groupby('Category')['Calcium (% Daily Value)'].mean()
```

```python
plt.figure(figsize=(20, 4), dpi=100)

category = iron.index
Iron = iron.values
Calcium = calcium.values

X_axis = np.arange(len(category))

plt.bar(X_axis - 0.2, Iron, 0.4, label = 'Iron', color = 'green')
plt.bar(X_axis + 0.2, Calcium, 0.4, label = 'Calcium', color = 'blue')

plt.xticks(X_axis, category)
plt.xlabel("Menu Categories")
plt.title("Distribution of Iron and Calcium in McDonald's Foods")
plt.xticks(rotation=0)
plt.legend()
plt.show()
```
